1...  MEMEORY EFFICIENCY:
 generators produce values one at a time instead of generating and storing the entire sequence in memory like regular functions. This is especially beneficial when dealing with large datasets or infintite sequence , as it prevent memory overflow.

2.. LAZY EVALUATION:
Generators evalutes values when requested ,This "lazy" evaluation can improve performance by avoiding unnecessary computations, especially when you only need a subset of the generting values.

Genertors can often simplify code for tasks involving iteration  . They encapsulate the logic of generating values , making the code more readble, and concise compared to equivalent implementations using regular function and loops.

10. What is diffrence between map(), reduce(), and filter(),function in python?


> 1. map()

: Applies a function to each element of an iterable (e.g., list, tuple) and returns a new iterable with the transformed elements.
Syntax: map(function, iterable)
Example:
```
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
```
In this example, map() applies the lambda function (which squares a number) to each element of the numbers list. The result is a new list, squared_numbers, containing the squared values.

2. reduce()

 Applies a function cumulatively to the elements of an iterable, reducing them to a single value.
Syntax: reduce(function, iterable[, initial])
Requires Import: from functools import reduce


Example:
```
from functools import reduce
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
```
Here, reduce() multiplies the elements of the numbers list step-by-step. It starts with the first two elements, applies the lambda function (multiplication), then takes the result and applies the function again with the next element, and so on, until it produces a single value (the product of all elements).

3. filter()

: Filters elements from an iterable based on a function that returns True or False for each element. It returns a new iterable containing only the elements for which the function returned True.
Syntax: filter(function, iterable)

Example:
```
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

```

````python
#create an iterator class in python that generates the fibonacci sequence up to a specified numbers of terms?
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_index = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_index >= self.num_terms:
            raise StopIteration

        if self.current_index == 0:
            self.current_index += 1
            return self.a
        elif self.current_index == 1:
            self.current_index += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.current_index += 1
            return next_value

# Example usage:
num_terms = 10
fib_iterator = FibonacciIterator(num_terms)

for fib_number in fib_iterator:
    print(fib_number)
````

````python
#write a generator function in python that yeilds the power of 2 up to a given exponent?
def powers_of_two(max_exponent):
    """Generator function that yields powers of 2 up to the given exponent."""
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

max_exponent = 5
for value in powers_of_two(max_exponent):
   print (value)
````

````python
#Implement  a generator function that reads a file line by line and yeilds each line as a string?
def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line and yields each line as a string.

    :param file_path: Path to the file to be read.
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


````

````python
#write a python program that uses `map()`to convert a list of tempratures from celsius to farenheit?
def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40, 100]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
````
